# Veri Filtreleme

The Dark Knight	filminin IMDB rating'inin dataframedeki sırasını biliyorsak eğer yerini bulalım;

```python
df["IMDB_Rating"][2]
```

Ancak dataframedki yerini bilmiyorsak;

```python
df.loc[df["Series_Title"]=="The Dark Knight"]["IMDB_Rating"]
```

Bu kodun algoritması şöyledir; öncelikle df.loc[df['Series_Title']=="The Dark Knight"] fonksiyonuyla satır bulunur. loc fonksiyonu lokasyondan türemiştir.

```python
df.loc[df['Series_Title']=="The Dark Knight"]
```

Ancak biz filmin IMDB puanını istemekteyiz. Bunun için IMDB_Rating sütununu seçiyoruz. 
Ancak öncelikle yukarıdaki fonksiyonun dönderdiği dataframein tipine bakalım;

```python
type(df.loc[df['Series_Title']=="The Dark Knight"])
```

O zaman bir dataframe'in basitçe sütununu seçmek için köşeli parantez kullanıyor ve seçiyoruz:

```python
df.loc[df["Series_Title"]=="The Dark Knight"]["IMDB_Rating"]
```

Son olarak da IMDB_Rating değeri 8'in üstünde olan ve No_of_Votes 100000den çok olan filmleri görelim:

```python
df.loc[(df['IMDB_Rating']>=8) & (df['No_of_Votes']>=100000)]
```

Şimdi bir de IMDB_Rating değeri 8'in üstünde olan ve Gross değeri (yani gişe hasılatı) 30000000'dan çok olan filmleri görelim:

```python
df.loc[(df['IMDB_Rating']>=8) & (df['Gross']>=30000000)]
```

Object tipinde olduğu için hata veriyor:

```python
df.dtypes
```

```python
# convert column "Gross" of a DataFrame
# to_numeric This function will try to change non-numeric 
# objects (such as strings) into integers or floating point numbers as appropriate.
df['Gross'] = df['Gross'].str.replace(",", "")
df["Gross"] = pd.to_numeric(df["Gross"])
type(df["Gross"][0])
```

```python
df.loc[(df['IMDB_Rating']>=8.0) & (df['Gross']>=300000000)]
```
